```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

path = 'data/'
```

```python
# Setting the path and reading the dataset
data = pd.read_csv(path + 'data.csv')
data_by_genres = pd.read_csv(path + 'data_by_genre_clean.csv')
```

```python
print(data.columns, '\n', data_by_genres.columns)
```

```python
# Finding names of all the unique artists
uniqueArtists = data.Artists.unique().tolist()
print(len(uniqueArtists))
```

```python
# Finding out top 10% popular songs
data[data["Popularity"] > 90]["Name"]

# Plotting popular songs
plt.figure(figsize=(16, 4))
plt.title('Top 15 Popular Songs', y = 1.25)
plt.xticks(rotation = 90)
x = data.groupby("Name")["Popularity"].mean().sort_values(ascending=False).head(15)
ax = sns.barplot(x.index, x)
ax.set_ylabel('Popularity')
ax.set_xlabel('Tracks')
```

```python
# Plotting a graph comparing Key
plt.figure(figsize=(16, 4))
plt.title('Key count comparison', y = 1.05)
plt.xlabel('Key')
plt.ylabel('Count')
sns.countplot(data['Key'])
plt.show()
```

```python
# Checking audio characteristics over years
fig,ax = plt.subplots(figsize=(30, 8))
sns.despine(fig, left=True, bottom=True)
sns.set_context(rc={"lines.linewidth": 3})

features = ["Acousticness","Danceability","Energy","Speechiness","Liveness","Valence"]
ax.set_ylabel('Measure')
ax.set_xlabel('Year')
ax.set_title('Audio trait trend over years', y = 1.05)
for col in features:
    x = data.groupby("Year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
# fig.savefig("viz/trends.png") 
```

```python
# Plotting popularity with different features
plt.figure(figsize=(20, 15))
plt.title('Popularity trend over different features', y = 1.05)
Xfeatures = ["Danceability", "Acousticness", "Loudness", "Energy", "Tempo", "Liveness", "Instrumentalness", "Valence","Year"] # Can add more features in this list
Yfeature = "Popularity"
cols = 4
rows = len(Xfeatures)//cols + 1
for idx, feat in enumerate(Xfeatures):
    plt.subplot(rows,cols,idx+1)
    sub_group = data.groupby(feat)
    Yfeature = sub_group.mean()["Popularity"] # Mean popularity is good enough to estimate and visualize
    sns.scatterplot(x=Yfeature.index, y=Yfeature)
```

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
pre90s = data[data['Year'] < 2000][["Danceability", "Acousticness", "Energy", "Liveness", "Instrumentalness", "Valence"]].describe()
post90s = data[data['Year'] >= 2000][["Danceability", "Acousticness", "Energy", "Liveness", "Instrumentalness", "Valence"]].describe()
```

```python
categories = ["Danceability", "Acousticness", "Energy", "Liveness", "Instrumentalness", "Valence"]

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=pre90s.loc['mean'],
      theta=categories,
      fill='toself',
      name='Pre 2000s'
))
fig.add_trace(go.Scatterpolar(
      r=post90s.loc['mean'],
      theta=categories,
      fill='toself',
      name='Post 2000s'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 0.8]
    )),
  showlegend=False
)

fig.show()
```
