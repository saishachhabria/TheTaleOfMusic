```python
import pandas as pd
import numpy as np
import re

path = 'data/'
```

```python
df = pd.read_csv(path + "data.csv")
print(df.shape)
df.head(5)
```

```python
ids = df["Id"].drop_duplicates()
```

```python
df_art = pd.read_csv(path + "data_w_genres.csv")
print(df_art.shape)
df_art.head(5)
```

```python
df_art.isnull().any()
```

```python
def get_genre(x):
    np.random.seed(42)
    lis = x[2:-2].split("', '")
    to_return = lis[np.random.randint(0, len(lis))]
    if to_return != '' and to_return.find('\\') == -1 and to_return.find('\'') == -1:
        return to_return
    return np.nan
df_art["genre"] = df_art["genres"].apply(lambda x: get_genre(x))
df_art = df_art.dropna()
print(df_art.shape)
df_art.head(5)
```

```python
genres = np.asanyarray(df_art["genre"].drop_duplicates().dropna())
# print("we are considering", genres.shape[0], "music genres")
# genres[:15]
```

```python
vc = df_art["genre"].value_counts()
df_vc = pd.DataFrame(vc).reset_index().rename(columns = {"genre": "count"})
df_vc = df_vc.rename(columns = {"index": "genre"})
labels = np.arange(df_vc.shape[0])
df_vc["label"] = labels
df_vc.head(5)
```

```python
df_labels = df_vc[["genre", "label"]]
df_labels.to_csv(r"genres_labeled.csv", index = False, header = True)
```

```python
df_clean = df_art.merge(df_vc, how = "left", left_on = "genre", right_on = "genre").rename(columns = {"count_y":"count"})
df_clean.head(5)
```

## Just use labels that have 60 or more items, thus we only predict 53 different music genres
